````python
import os
import glob
voice='wukong-0.0.1-e300-rmvpe-cnhubert-SingerPreTrain'
def delete_except(retain_files):
    # 获取当前目录下的所有文件
    all_files = glob.glob(f'/kaggle/working/program_ml/logs/{voice}/*')
    all_files.extend(glob.glob(f'/kaggle/working/program_ml/logs/{voice}/**/*'))
    # 创建一个需要删除的文件列表
    files_to_delete = [file for file in all_files if file not in retain_files]
    
    # 删除不需要保留的文件
    for file in files_to_delete:
        try:
            os.remove(file)
            print(f"Deleted {file}")
        except OSError as e:
            print(f"Error: {e.strerror} - {file}")
# 指定要保留的文件名称
retain_files = glob.glob(f'/kaggle/working/program_ml/logs/{voice}/*{voice}*')
# delete_except(retain_files)
````

````python
import os

# 定义源目录和目标目录
src_dir = "/kaggle/input"
dst_dir = "/kaggle/working/program_ml/assets/datasets"

for subdir in os.listdir(src_dir):
    src_path = os.path.join(src_dir, subdir)
    dst_path = os.path.join(dst_dir, subdir)
    os.symlink(src_path, dst_path)
    print(f"Created symlink: {dst_path} -> {src_path}")
````

````python
from huggingface_hub import HfApi
voice='wukong-0.0.1-e300-rmvpe-cnhubert-SingerPreTrain'
api = HfApi()
import glob
models=glob.glob(f"/kaggle/working/program_ml/logs/{voice}/*{voice}*")
# !ls /kaggle/working/program_ml/logs/wukong-0.0.1-e300-rmvpe-cnhubert-SingerPreTrain
for f in models:
    fn=os.path.basename(f)
    api.upload_file(
        path_or_fileobj=f,
        path_in_repo=f"{voice}/fn",
        repo_id="ywaby/rvc-trained",
        repo_type="model",
    )
````
